# 1. Bifurcation diagram WITH parallelization

```python
%%writefile Nazarova_Parallel_Bifurcation_diagram.py

from mpi4py import MPI
import numpy as np
import matplotlib.pyplot as plt

# Initialization for parallelization
comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

# Initialization for Bifurcation diagram
steps = 500
n = [500, 330, 250, 200, 160, 140, 120]
ran = np.linspace(0, 4, size*n[size-2])

def bifurcation_map(r, x):
    return r*x*(1-x)

# Initialization for who takes which piece of r
distr = len(ran)//(size)

list_of_distr = list(range(0, len(ran), distr))
list_of_distr.append(len(ran))

list_to_send = []
for i in range(len(list_of_distr)-1):
    list_to_send.append([list_of_distr[i],list_of_distr[i+1]])


# Parallelization itself
sendbuf = None
Rs_send = None
if rank == 0:
    sendbuf = np.empty([size*n[size-2], steps], dtype='f')
    sendbuf.T[:,:] = range(size*n[size-2])

    Rs_send = np.empty([size*n[size-2], steps], dtype='f')
    Rs_send.T[:,:] = range(size*n[size-2])

recvbuf = np.empty((n[size-2], steps), dtype='f')
comm.Scatter(sendbuf, recvbuf, root=0)

Rs_res = np.empty((n[size-2], steps), dtype='f')
comm.Scatter(Rs_send, Rs_res, root=0)

list_to_send = list_to_send[rank]

j = -1
for r in ran[list_to_send[0]:list_to_send[1]]:
    j += 1
    x = np.random.random()
    for i in range(steps):
        recvbuf[j, i] = x
        Rs_res[j, i] = r
        x = bifurcation_map(r, x)

comm.Barrier()

comm.Gather(recvbuf, sendbuf, root=0)
comm.Gather(Rs_res, Rs_send, root=0)

if rank == 0:
    Rs_send = Rs_send[:,Rs_send.shape[1]//2:]
    sendbuf = sendbuf[:,sendbuf.shape[1]//2:]
    Rs_send = Rs_send.reshape(Rs_send.shape[0]*Rs_send.shape[1])
    sendbuf = sendbuf.reshape(sendbuf.shape[0]*sendbuf.shape[1])
    plt.figure(figsize=(9, 8))
    plt.title('Bifurcation map')
    plt.xlabel("r")
    plt.ylabel("x")
    plt.plot(Rs_send, sendbuf, 'r.', markersize=1)
    plt.grid()
    plt.pause(1)
            
```

```python
%timeit !mpirun --allow-run-as-root -np 8 python3 Nazarova_Parallel_Bifurcation_diagram.py
```

# 2. Plot speedup vs number of processes

```python
import matplotlib.pyplot as plt

n_CPU = list(range(1, 9))
#1 cpu: 4.42 s ± 269 ms per loop; 2 cpu: 1.8 s ± 150 ms per loop; 3 cpu: 1.86 s ± 118 ms per loop; 4 cpu: 1.85 s ± 90.1 ms per loop; 5 cpu: 2.11 s ± 41.3 ms per loop; 
# 6 cpu: 2.38 s ± 112 ms per loop; 7 cpu: 2.91 s ± 180 ms per loop; 8 cpu: 3.36 s ± 124 ms per loop
speedup = [4.42, 1.8, 1.86, 1.85, 2.11, 2.38, 2.91, 3.36]
```

```python
plt.figure(figsize=(12, 8))
plt.plot(n_CPU, speedup, c='orange')
plt.title('Speedup vs number of Processes')
plt.xlabel('Number of Processes')
plt.ylabel('Speedup, s per loop')
plt.show()
```

Conclusion:

The plot showing the dependence of the speedup VS number of processes shows that when we use only 1 process, the speedup is high. Processes 2 and 3 coped with the task best of all, there is also a logical explanation for this (This is more my assumption, but I will still share it). This is because the algorithm is not large enough to parallelize. Processes spend more time on commands such as: comm.Scatter, comm.Gather and when they expect each other comm.Barrier().

### 3.1 Implement the map, plot the evolution of x

```python
def bifurcation_map(r, x):
    return r*x*(1-x)

x = np.linspace(0, 1, 100)

plt.plot(x, bifurcation_map(2, x))
plt.show()
```

```python
def evolution(x_0, r, steps = 200):
    points = np.array([[x_0, x_0]])
    plt.figure(figsize=(9, 8))
#     plt.subplot(2, 2, 1)
    
    for i in range(0, steps//2):
        points = np.append(points, [[points[-1][0], bifurcation_map(r, points[-1][0])]], axis=0)
        points = np.append(points, [[points[-1][1], points[-1][1]]], axis=0)
        
    x = np.linspace(0, 1, 100)
    plt.plot(x, x, c ='orange')
    plt.plot(x, bifurcation_map(r, x))
    
    for i in range(points.shape[0]):
        plt.scatter(points[i][0], points[i][1], c ='r')
        
    plt.plot(points[:,0], points[:,1], c = 'r', linewidth=.4)
    plt.title('Evolution map_1')
    plt.xlabel("x")
    plt.ylabel("res of logistic func")
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(9, 8))
    n = np.arange(0, steps+1, 1)
    plt.plot(n, points[:,0])
    plt.title('Evolution map_2')
    plt.xlabel("steps")
    plt.ylabel("fish")
    plt.grid()
    plt.show()
#     print(points)
#     print(n)

#     return points
```

```python
evolution(0.1, 3)
```

### 3.2 Create a linspace of r’s, for every r save last “m” values of x after first “n” values (can be m=200, x=200), play around with values

```python
def evolution_2(x_0, steps = 500):
    ran = np.linspace(0, 4, 1000)
    result_0 = np.zeros(((steps - steps//2 +1),2))
    result = np.array([[0,0]])

    for r in ran:
        points = np.array([[x_0, x_0]])
        for i in range(0, steps//2):
            points = np.append(points, [[points[-1][0], bifurcation_map(r, points[-1][0])]], axis=0)
            points = np.append(points, [[points[-1][1], points[-1][1]]], axis=0)
        
        result_0[:,0] = r
        result_0[:,1] = points[:,0][steps//2:]
        result = np.append(result, result_0, axis = 0)
          
    plt.figure(figsize=(9, 8))
    plt.title('Bifurcation map')
    plt.xlabel("r")
    plt.ylabel("x")

    plt.plot(result[:,0], result[:,1], 'r.', markersize=1)

    plt.grid()
    plt.show()

#     return result
```
