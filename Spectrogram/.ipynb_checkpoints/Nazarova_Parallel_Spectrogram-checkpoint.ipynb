# 1. Spectrogram WITH parallelization

```python
%%writefile Nazarova_Parallel_Spectrogram.py

from mpi4py import MPI
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
from numpy import exp
from numpy import linspace
from numpy import fft

# Initialization for parallelization
comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

# Initialization for Spectrogram
n = 1000

t = linspace(-20*2*pi, 20*2*pi, n)
y = np.sin(t)*exp(-t**2/2/20**2)
y = y+np.sin(3*t)*exp(-(t-5*2*pi)**2/2/20**2)
y = y+np.sin(5.5*t)*exp(-(t-10*2*pi)**2/2/5**2)
y = y+np.sin(4*t)*exp(-(t-7*2*pi)**2/2/5**2)
y = np.array(y)


# Initialization for who takes which piece of w
distr = n//(size)
list_of_distr = list(range(0, n+1, distr))
if list_of_distr[-1] != n:
    list_of_distr[-1] = n
    
list_to_send = []
for i in range(len(list_of_distr)-1):
    list_to_send.append([list_of_distr[i],list_of_distr[i+1]])
    
list_to_send = list_to_send[rank]


# Parallelization itself
windows = linspace(-20*2*pi,20*2*pi,n)
Res = None

specgram = []
for i in range(list_to_send[0], list_to_send[1]):
    window_function = exp(-(t-windows[i])**2/2/(2.0*2*pi)**2)
    y_window = y * window_function
    y_windowed = np.array(y_window)
    specgram.append(abs(fft.fft(y_windowed)))

specgram = np.array(specgram)
comm.Barrier()

Res = comm.gather(specgram, root=0)

if rank == 0:
    Res = np.array(np.concatenate(Res).ravel())
    Res = Res.reshape((n, n), order='F')
    
    t = np.linspace(-20*2*pi, 20*2*pi, n)
    w = fft.fftfreq(len(y), d=(t[1]-t[0])/2/pi)

    plt.figure(figsize=(10,7))
    plt.imshow(Res, aspect='auto', extent=[-20, 20, y[0], 2 * w[int(len(t)/2)-1]], cmap='plasma')
    plt.ylim(0,10)
    plt.title("Spectogram")
    plt.xlabel('T,cycles')
    plt.ylabel('Frequency')
    plt.colorbar()
    plt.show()
```

### 3.1. Add 4th wave packet (frequency = 4 and time_shift = 7 cycles). Demonstrate the effect on the plot of the FFT spectrum

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

## Initial Signal
nofts = 1000
t = np.linspace(-20*2*pi, 20*2*pi, 2**14+5)
t = np.linspace(-20*2*pi, 20*2*pi, 3**9+6)

y = np.sin(t)*exp(-t**2/2/20**2)
y = y+np.sin(3*t)*exp(-(t-5*2*pi)**2/2/20**2)
y = y+np.sin(5.5*t)*exp(-(t+10*2*pi)**2/2/5**2)

# Initial Fourier spectrum
sp = np.fft.fft(y)
w = fft.fftfreq(len(y), d=(t[1]-t[0])/2/pi)

## Signal with 4th wave packet
y_4th_wave = y+np.sin(4*t)*exp(-(t-7*2*pi)**2/2/5**2)

# Fourier spectrum with 4th wave packet
sp_4th_wave = np.fft.fft(y_4th_wave)
w_4th_wave = fft.fftfreq(len(y_4th_wave), d=(t[1]-t[0])/2/pi)

## Initial Signal ploting
axs[0, 0].plot(t/2/pi,y)
axs[0, 0].set_title('Initial Signal')
axs[0, 0].set_xlabel('t, cycles')
axs[0, 0].set_ylabel('signal, arb.units')

# Initial Fourier spectrum ploting
axs[0, 1].plot(w, abs(sp)**2)
axs[0, 1].set_title('Initial Fourier spectrum')
axs[0, 1].set_xlabel('Frequency, arb. units')
axs[0, 1].set_ylabel('Power spectrum, arb. units')
axs[0, 1].set_xlim(0,6)

## Signal with 4th wave packet ploting
axs[1, 0].plot(t/2/pi,y_4th_wave, color='orange')
axs[1, 0].set_title('Signal with 4th wave packet')
axs[1, 0].set_xlabel('t, cycles')
axs[1, 0].set_ylabel('signal, arb.units')

# Fourier spectrum ploting with 4th wave packet ploting
axs[1, 1].plot(w_4th_wave, abs(sp_4th_wave)**2, color='orange')
axs[1, 1].set_title('Fourier spectrum with 4th wave packet ploting')
axs[1, 1].set_xlabel('Frequency, arb. units')
axs[1, 1].set_ylabel('Power spectrum, arb. units')
axs[1, 1].set_xlim(0,6)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(14, 10))

# windowed Fourier transform (Gabor transform, spectrogram)

window_width=2.0*2*pi
window_position=-5.0*2*pi
window_function=exp(-(t-window_position)**2/2/window_width**2)

y_window=y * window_function

axs[0].plot(t/2/pi, y, color='m')
axs[0].plot(t/2/pi, y_window, color='c')
axs[0].set_title('Initial Windowed Fourier transform')
axs[0].set_xlabel('Window position, cycles')
axs[0].set_ylabel('frequency, arb. units')

fig.tight_layout(pad=3.0)

# windowed Fourier transform (Gabor transform, spectrogram) with 4th wave
y_window_4th_wave = y_4th_wave * window_function

axs[1].plot(t/2/pi, y_4th_wave, color='c')
axs[1].plot(t/2/pi, y_window_4th_wave, color='m')
axs[1].set_title('Windowed Fourier transform with 4th wave packet ploting')
axs[1].set_xlabel('Window position, cycles')
axs[1].set_ylabel('frequency, arb. units')
```

### 3.2. Implement the spectrogram, show the effect of (1) on the spectrogram. Don’t forget to label the axes

```python
def get_specgram(n=10000):

    t = np.linspace(-20*2*pi, 20*2*pi, n)
    y = np.sin(t)*exp(-t**2/2/20**2)
    y = y+np.sin(3*t)*exp(-(t-5*2*pi)**2/2/20**2)
    y = y+np.sin(5.5*t)*exp(-(t-10*2*pi)**2/2/5**2)
    y = y+np.sin(4*t)*exp(-(t-7*2*pi)**2/2/5**2)
    y = np.array(y)

    windows = linspace(-20*2*pi,20*2*pi,n)

    specgram=np.zeros((n,n))
    for n,position in enumerate(windows):
        window_function=exp(-(t-position)**2/2/(2.0*2*pi)**2)
        y_window = y * window_function
        y_windowed = np.array(y_window)
        specgram[:,n]=abs(fft.fft(y_windowed))

    t = np.linspace(-20*2*pi, 20*2*pi, n)
    w = fft.fftfreq(len(y), d=(t[1]-t[0])/2/pi)

    plt.figure(figsize=(10,7))
    imshow(specgram, aspect='auto', extent=[-20, 20, y[0], 2 * w[int(len(t)/2)-1]], cmap='plasma')
    plt.ylim(0,10)
    plt.title("Spectogram")
    plt.xlabel('T,cycles')
    plt.ylabel('Frequency')
    colorbar()
```

```python
sp = get_specgram()

# %timeit sp = get_specgram()
```
